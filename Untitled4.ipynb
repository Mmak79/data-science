```python
params = {
      'apiid' : 'ccf98e64-0bd8-4cb3-8c12-ff6411d239d4',
      'lang': 'ru'
}
url = 'https://api.delivery-club.ru/api1.2/vendors/search?latitude=56.35&longitude=37.5167&query=%D0%BA%D1%83%D1%80%D0%B8%D0%BD%D1%8B%D0%B5+%D0%BA%D1%80%D1%8B%D0%BB%D1%8B%D1%88%D0%BA%D0%B8&cacheBreaker=1665170550'

response = requests.get(url=url, params=params)

print(response.status_code)
pprint(response.json())
```

```python
vendors = response.json()["vendors"]
```
